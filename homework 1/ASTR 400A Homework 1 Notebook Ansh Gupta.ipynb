```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n0_polytope = lambda x: 1 - x**2/6
n1_polytope = lambda x: np.sin(x)/x
n5_polytope = lambda x: (1 + x**2/3)**(-1/2)
```

```python
s = 1000
x_max = 5
xi_0 = np.hstack((np.linspace(1e-5, np.sqrt(6), s), [np.sqrt(6)+0.01, x_max]))
xi_1 = np.hstack((np.linspace(1e-5, np.pi, s), [np.pi+0.01, x_max]))
xi_5 = np.linspace(0, x_max, s)

rho_0 = np.hstack(((n0_polytope(xi_0)**0)[:-2], [0, 0]))
rho_1 = np.hstack(((n1_polytope(xi_1)**1)[:-2], [0, 0]))
rho_5 = n5_polytope(xi_5)**5

# norm_0 = np.trapz(rho_0, x=xi_0)
# norm_1 = np.trapz(rho_1, x=xi_1)
# norm_5 = np.trapz(rho_5, x=xi_5)

norm_0 = rho_0[0]
norm_1 = rho_1[0]
norm_5 = rho_5[0]
```

```python
plt.figure(figsize=(8, 6))
plt.plot(xi_0, rho_0/norm_0, c='black', label='n = 0', zorder=1)
plt.plot(xi_1, rho_1/norm_1, c='#DC267F', label='n = 1', zorder=2)
plt.plot(xi_5, rho_5/norm_5, c='#FFB000', label='n = 5', zorder=3)

plt.title("Density Profile of Stars for Polytropic Index n = 0, 1, and 5")
plt.xlabel(r'$\xi = r/\lambda_{n}$')
plt.ylabel(r'$\rho/\rho_c$')
plt.legend()
plt.figtext(0.3, 0.01, r'At $r = 0$, the density $\rho$ is equal to the central density $\rho_0$')
plt.savefig('homework 1/density-profiles.png', dpi=400)
```
