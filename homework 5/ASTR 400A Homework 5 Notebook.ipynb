```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import astropy.units as u
import astropy.constants as c
import mesa_web
```

```python
low_metal_dir = 'low-metal-low-mass/'
low_metal_subdir = 'MESA-Web_Job_09262318630/'
high_metal_dir = 'high-metal-low-mass/'
high_metal_subdir = 'MESA-Web_Job_09262318628/'

low_metal_history_fname = low_metal_dir + low_metal_subdir + 'trimmed_history.data'
low_metal_data_dict = mesa_web.read_history(low_metal_history_fname)
high_metal_history_fname = high_metal_dir + high_metal_subdir + 'trimmed_history.data'
high_metal_data_dict = mesa_web.read_history(high_metal_history_fname)
```

```python
low_metal_age = low_metal_data_dict['star_age']
low_metal_Teff = 10**low_metal_data_dict['log_Teff']
low_metal_L = 10**low_metal_data_dict['log_L']
low_metal_L_pp = 10**low_metal_data_dict['pp']
low_metal_L_cno = 10**low_metal_data_dict['cno']
```

```python
high_metal_age = high_metal_data_dict['star_age']
high_metal_Teff = 10**high_metal_data_dict['log_Teff']
high_metal_L = 10**high_metal_data_dict['log_L']
high_metal_L_pp = 10**high_metal_data_dict['pp']
high_metal_L_cno = 10**high_metal_data_dict['cno']
```

```python
plt.semilogx(low_metal_age, low_metal_Teff, c='blue', ls='--', label='Z = 0.0001')
plt.semilogx(high_metal_age, high_metal_Teff, c='red', label='Z = 0.02')

plt.title(r'Effective Temperature vs Age for High and Low Metallicity')
plt.xlabel(r' Age [yr] ')
plt.ylabel(r' T$_{eff}$ [K]')
plt.xlim(9e5, 2e7)
plt.ylim(1e4, 3e4)
plt.legend()

print(f'Effective temperature of the low-metallicity star is approximately {23000/17000:.3} \
times that of the high-metallicity star')
```

```python
plt.semilogx(low_metal_age, low_metal_L, c='blue', ls='--', label='Z = 0.0001')
plt.semilogx(high_metal_age, high_metal_L, c='red', label='Z = 0.02')

plt.title(r'Luminosity vs Age for High and Low Metallicity')
plt.xlabel(r' Age [yr] ')
plt.ylabel(r' Luminosity [L$_{\odot}$]')
plt.xlim(9e5, 2e7)
plt.ylim(4e2, 1e3)
plt.legend()

print(f'Luminosity of the low-metallicity star is approximately {710/540:.3} \
times that of the high-metallicity star')
```

The high-metallicity star has a lower effective temperature than the low-metallicity star. We also see that the high-metallicity star appears less luminous. How can we explain these facts?

As I describe in part b), the low-metallicity star has a lower opacity because it has fewer free electrons. That means we can see further into the low-metallicity star than the high-metallicity one, since  $\frac{2}{3}\tau$ is a larger distance when the opacity is lower. The deeper regions of the low-metallicity star we observe will be hotter since these stars have the same mass. In contrast, we see light emitted closer to the (cooler) surface in the high-metallicity star. This directly explains why the high-metallicity star has a lower effective temperature than the low-metallicity one. Since luminosity has a T$^4$ dependence, this also explains why the high-metallicity star is less luminous.

I speculate that the luminosity gain from the higher temperature of the emitting surface of the low-metallicity star is probably somewhat cancelled out by the reduction in visible surface area.

## b)
Stars with more metals will have a greater number of free electrons (since the neutral versions of those metals came along with tons of electrons, which inevitably get stripped as everything heats up and turns into a plasma). These electrons increase opacity in multiple ways. For one, the free electrons can combine with neutral hydrogren to form H$^{-}$ ions, which can increase the opacity by allowing for free-free absorption (and bound-free emission at low photon energies). However, at these high temperatures, it's likely that most of the electrons will be free rather than in H$^{-}$, which still allows for free-free absorption but also introduces significant electron scattering component.

```python
plt.semilogx(low_metal_age, low_metal_L_pp, c='blue', ls='--', label='Z = 0.0001')
plt.semilogx(high_metal_age, high_metal_L_pp, c='red', label='Z = 0.02')

plt.title(r'p-p Luminosity vs Age for High and Low Metallicity')
plt.xlabel(r' Age [yr] ')
plt.ylabel(r'p-p Luminosity [L$_{\odot}$]')
plt.xlim(9e5, 2e7)
plt.ylim(0, 150)
plt.legend()

print(f'Luminosity of the low-metallicity star is approximately {710/540:.3} \
times that of the high-metallicity star')
```

```python
plt.semilogx(low_metal_age, low_metal_L_cno, c='blue', ls='--', label='Z = 0.0001')
plt.semilogx(high_metal_age, high_metal_L_cno, c='red', label='Z = 0.02')

plt.title(r'CNO Luminosity vs Age for High and Low Metallicity')
plt.xlabel(r' Age [yr] ')
plt.ylabel(r'CNO Luminosity [L$_{\odot}$]')
plt.xlim(1e6, 2e7)
plt.ylim(450, 800)
plt.legend()

print(f'Luminosity of the low-metallicity star is approximately {710/540:.3} \
times that of the high-metallicity star')
```

We can see that during the main sequence phase, the lower-metallicity star has a much greater luminosity produced by both the p-p and CNO processes. It's not a perfect comparison because other physical conditions in these stars may differ due to their compositional differences, but this loosely suggests that the p-p and CNO reaction rates in the low-metallicity star are higher than in the high-metallicity star.

At a basic level, this seems to be because the low-metallicity star is slightly hotter (throughout) than the high-metallicity one, as seen in the plot below. To explain this, I tried thinking about the balance between radiation pressure and gravity. The radiation in the low-metallicity (low-opacity) star has a harder time "pushing" against the material, because the opacity is lower. As a result, the star has to "burn brighter" to provide the same pushing force against gravity.

Echoing this, we can also see that the p-p process energy generation rate is higher in the low-metallicity star at the specific time step I chose, which is towards the tail end of the main sequence (plots shown below). However, the CNO process actually has a slightly higher energy generation rate in the high-metallicity star. This goes against my intuition, especially since the total CNO luminosity is much greater in the low-luminosity star. My guess is that because the higher metallicity includes the catalyst elements, making the overall reaction more efficient. However, the CNO luminosity of the low-metallicity star is probably still much greater because the amount of mass at the required temperature is greater.
